```python
db = sa.create_engine('sqlite:///db_olist.sqlite2',echo = False)
conn = db.connect()
```

```python
# 01. Quantos pedidos foram feitos para cada tipo de pagamento?

query = """
SELECT 
    payment_type,
    order_id,
    COUNT( order_id ) as payment_type
FROM order_payments
GROUP BY payment_type
ORDER BY payment_type DESC;
"""

df = pd.read_sql_query( query, conn )
df.head()
```

```python
## 02. Qual o número máximo e mínimo de parcelas nos pagamentos?

query = """
SELECT 
    
    MIN(payment_installments),
    MAX(payment_installments)
    
FROM order_payments

"""

df = pd.read_sql_query( query, conn )
df.head()
```

```python
## 03. Quais são os top 10 pedidos com os maiores valores?

query = """
SELECT 
	order_id,
	payment_value		
FROM order_payments
ORDER BY payment_value DESC 
LIMIT (10);
"""

df = pd.read_sql_query( query, conn )
df.head(10)

```

```python
#04. Quais são os últimos 10 pedidos com os menores valores?

query = """
SELECT 
	order_id,
	payment_value		
FROM order_payments
WHERE payment_value > 0
ORDER BY payment_value ASC;
"""

df = pd.read_sql_query( query, conn )
df.head(10)
```

```python
# 05. Qual a média do valor de pagamento por tipo de pagamento?

query = """
SELECT 
    payment_type,
    
    AVG ( payment_value ) AS mean_value
    
FROM order_payments
GROUP BY payment_type
ORDER BY mean_value DESC  ;
"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 06. Quais os top 5 clientes com os maiores valores de pagamento no boleto?


query = """
SELECT 

    payment_type,
    order_id,
    payment_value
   
    
FROM order_payments op 
WHERE 
(payment_type='boleto')

--GROUP BY payment_type ="boleto"
ORDER BY payment_value DESC  
LIMIT (5);
"""

df = pd.read_sql_query( query, conn )
df

```

```python
# 07. Quais os top 5 clientes com os maiores valores de pagamento no cartão de crédito?

query = """
SELECT 

    payment_type,
    order_id,
    payment_value
   
    
FROM order_payments op 
WHERE 
(payment_type='credit_card')

--GROUP BY payment_type ="boleto"
ORDER BY payment_value DESC  
LIMIT (5);
"""

df = pd.read_sql_query( query, conn )
df

```

```python
# 08. Quais os 10 produtos mais caros?

query = """
SELECT 
	price,
	product_category_name
	
FROM order_items oi inner join products p on (oi.product_id = p.product_id)

--GROUP BY price
ORDER BY price DESC  
LIMIT (10);
"""

df = pd.read_sql_query( query, conn )
df

```

```python
# 09. Quais os 10 produtos mais baratos?

query = """
SELECT 
	price,
	product_category_name
	
FROM order_items oi inner join products p on (oi.product_id = p.product_id)

--GROUP BY price
ORDER BY price ASC  
LIMIT (10);
"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 10. Quais as 10 categorias mais compradas?

query = """
SELECT 
	product_category_name,
	count(order_id) as num_orders
FROM order_items oi inner join products p on (oi.product_id = p.product_id)

GROUP BY product_category_name
ORDER BY num_orders DESC
LIMIT (10);
"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 11. Quais os 5 produtos com maior número de reviews?

query = """SELECT
    oi.product_id,
    COUNT(orv.review_score) AS num_reviews
    
FROM order_items oi JOIN order_reviews orv ON (oi.order_id = orv.order_id)
GROUP BY oi.product_id
ORDER BY num_reviews DESC
LIMIT(5);"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 12. Quais os top 10 produtos sem nenhum review?

query = """SELECT
    oi.product_id,
    COUNT(orv.review_score) AS num_reviews
    
FROM order_items oi JOIN order_reviews orv ON (oi.order_id = orv.order_id)
WHERE orv.review_comment_message IS NULL
GROUP BY oi.product_id
ORDER BY num_reviews DESC
LIMIT(10);"""

df = pd.read_sql_query( query, conn )
df

```

```python
# 13. Quais os 10 clientes com maior quantidade de pedidos?

query = """SELECT
    c.customer_id,
    COUNT(DISTINCT o.order_id ) AS num_orders
    
FROM orders o  JOIN customer c  ON (o.customer_id = c.customer_id  )

GROUP BY c.customer_id

ORDER BY num_orders DESC

LIMIT(10);"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 14. Quais os 10 clientes com a menor quantidade de pedidos?
query = """

SELECT
    c.customer_id,
    COUNT(DISTINCT o.order_id ) AS num_orders
    
FROM orders o  JOIN customer c  ON (o.customer_id = c.customer_id  )

GROUP BY c.customer_id

ORDER BY num_orders

LIMIT(10);"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 15. Quais vendedores existem na base?

query = """

SELECT
   DISTINCT seller_id
FROM sellers
LIMIT (10)
"""

df = pd.read_sql_query( query, conn )
df

```

```python
# 16. Qual a distribuição de vendedores por estado?

query = """

SELECT
   seller_state,
   COUNT(seller_id) AS num_sellers
FROM sellers
GROUP BY seller_state
ORDER BY num_sellers DESC

"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 17. Qual a distribuição de clientes por estado?

query = """

SELECT
   customer_state,
   COUNT(customer_id) AS num_customers
FROM customer c
GROUP BY customer_state
ORDER BY num_customers DESC;

"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 18. Quais são os top 10 vendedores que mais receberam pagamentos por boleto?

query = """

SELECT
   oi.seller_id,
COUNT(oi.order_id) AS num_orders
FROM order_items oi JOIN order_payments op ON (oi.order_id = op.order_id)
WHERE payment_type = 'boleto'
GROUP BY oi.seller_id
ORDER BY num_orders DESC
LIMIT 10;
"""

df = pd.read_sql_query( query, conn )
df
```

```python
# 19. Quais são os 10 piores vendedores em termos de número de vendas?

query = """

SELECT
   oi.seller_id,
COUNT(oi.order_id) AS num_orders
FROM order_items oi 

GROUP BY oi.seller_id
ORDER BY num_orders
LIMIT 10;
"""

df = pd.read_sql_query( query, conn )
df

```

```python
# 20. Quantos produtos são comprados, em média, por pedido?

query = """

WITH products_per_order AS (

SELECT
order_id,
COUNT(product_id) AS num_products
FROM order_items
GROUP BY order_id

)
SELECT AVG(num_products) AS avg_products_per_order
FROM products_per_order

"""

df = pd.read_sql_query( query, conn )
df


```
